# Prompt Engineering Assignment

This notebook demonstrates writing prompts for different purposes: resume, business idea, and study plan. For each, we compare a weak prompt with a strong prompt.

## Resume Prompts

### Weak Prompt
A weak prompt is vague, lacks specificity, and doesn't guide the AI towards producing a high-quality output.

### Strong Prompt
A strong prompt is detailed, specifies the format, includes context, and sets clear expectations.

```python
# Weak Resume Prompt
weak_resume_prompt = "Write a resume."

# Strong Resume Prompt
strong_resume_prompt = """Create a professional resume for a software engineer with 5 years of experience. Include the following sections:
- Contact Information
- Professional Summary
- Skills (technical and soft skills)
- Work Experience (with job titles, companies, dates, and bullet points describing achievements)
- Education
- Certifications (if any)

Use a clean, ATS-friendly format. Quantify achievements where possible, e.g., 'Improved system performance by 30%'. Tailor it for a mid-level software engineering role."""

print("Weak Resume Prompt:")
print(weak_resume_prompt)
print("\nStrong Resume Prompt:")
print(strong_resume_prompt)
```

### Comparison
The weak prompt is too vague; it doesn't specify the role, experience, or format, leading to generic or incomplete output.

The strong prompt provides specific details: role, experience level, required sections, format guidelines, and examples. This results in a more targeted and professional resume.

## Business Idea Prompts

```python
# Weak Business Idea Prompt
weak_business_prompt = "Give me a business idea."

# Strong Business Idea Prompt
strong_business_prompt = """Generate a detailed business idea in the sustainable energy sector. The idea should target small businesses and focus on renewable sources like solar or wind. Include:
- Problem it solves
- Target market
- Unique value proposition
- Revenue model
- Key features of the product/service
- Potential challenges and solutions
- Market size estimate
- Steps to launch

Ensure the idea is innovative, feasible, and environmentally friendly."""

print("Weak Business Idea Prompt:")
print(weak_business_prompt)
print("\nStrong Business Idea Prompt:")
print(strong_business_prompt)
```

### Comparison
The weak prompt lacks any constraints or details, potentially resulting in generic or irrelevant ideas.

The strong prompt specifies the industry, target audience, focus area, and required elements, leading to a comprehensive and targeted business idea.

## Study Plan Prompts

```python
# Weak Study Plan Prompt
weak_study_prompt = "Make a study plan."

# Strong Study Plan Prompt
strong_study_prompt = """Create a 4-week study plan for learning Python programming for beginners. The plan should include:
- Daily goals and topics
- Weekly milestones
- Recommended resources (books, websites, videos)
- Practice exercises
- Assessment methods
- Time allocation (e.g., 2 hours per day)
- Tips for staying motivated

Assume the learner has no prior programming experience. Structure it week by week with specific objectives."""

print("Weak Study Plan Prompt:")
print(weak_study_prompt)
print("\nStrong Study Plan Prompt:")
print(strong_study_prompt)
```

### Comparison
The weak prompt is ambiguous; it doesn't specify the subject, duration, or learner's level, resulting in a generic plan.

The strong prompt defines the subject, timeframe, learner's background, and required components, producing a structured and effective study plan.

## Conclusion

Strong prompts are specific, detailed, and provide context, leading to better, more relevant AI-generated outputs. Weak prompts are vague and often result in generic responses. Effective prompt engineering involves clearly defining the task, audience, format, and constraints.
