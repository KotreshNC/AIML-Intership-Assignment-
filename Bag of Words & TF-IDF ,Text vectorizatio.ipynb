```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import pandas as pd

# Dataset
documents = [
    "I love this movie",
    "This movie is terrible",
    "Amazing acting",
    "Worst film ever"
]

# 1) Bag of Words
bow = CountVectorizer()
bow_matrix = bow.fit_transform(documents)
bow_df = pd.DataFrame(bow_matrix.toarray(), columns=bow.get_feature_names_out())
print("Bag of Words features:\n", bow.get_feature_names_out())
print("\nBag of Words matrix:\n", bow_df)

# 2) TF-IDF
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(documents)
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out())
print("\nTF-IDF features:\n", tfidf.get_feature_names_out())
print("\nTF-IDF matrix:\n", tfidf_df)

# 3) Compare results
print("\nComparison:\n")
compare_df = pd.DataFrame({
    "term": bow.get_feature_names_out(),
    "bow_count": bow_df.sum(axis=0).values,
    "bow_df": bow_df.sum(axis=0).values,
})

print("\nBag of Words and TF-IDF preview (dense):")
print(pd.concat([bow_df, tfidf_df], axis=1))

# Extra: display both with pandas
comparison = pd.DataFrame(
    bow_df.T,
    index=bow_df.columns,
    columns=[f"doc_{i+1}" for i in range(len(documents))]
)
comparison["tfidf_average"] = tfidf_df.mean(axis=0).tolist()
print("\nExtra: Pandas comparison table (BoW counts + average TF-IDF):")
print(comparison)
```
